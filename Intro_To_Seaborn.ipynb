COUNT PLOT:

```python
var = sns.load_dataset("tips")
var
```

```python
sns.countplot(x="sex",data=var,hue="smoker")
plt.show()
```

PAIR PLOT:

```python
sns.pairplot(var,hue="sex")
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex")
plt.show()
```

```python
sns.pairplot(var,hue="sex",kind="reg") #'scatter' , 'kde' , 'hist' , 'reg'
plt.show()
```

STRIP PLOT:

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r")
plt.show()
```

BOX PLOT:

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="time",showmeans=True)
plt.show()
```
