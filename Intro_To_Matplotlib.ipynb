```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[2,4,6,8]
plt.plot(x,y)
plt.show()
```

```python
c=["r","y","g","b"]
plt.bar(x,y,color=c)
plt.show()
```

BAR PLOT

```python
x = ["python","c","c++","Java"]
y=[85,70,60,82]
plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("Programming Language",color="r")
plt.bar(x,y,label="popularity")
plt.legend()
plt.show()
```

```python
plt.barh(x,y,label="popularity") #horizontal bar
plt.legend()
plt.show()
```

SCATTER PLOT

```python
day = [1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=[10,49,30,29,56,20,30]
plt.scatter(day,no,c=colors,cmap="viridis")
plt.colorbar()
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

HISTOGRAM PLOT

```python
import numpy as np
import random

x=np.random.randint(10,60,(50))
x
```

```python
plt.hist(x,edgecolor="r" )
plt.xlabel("Programming")
plt.ylabel("No.")
plt.title("Histogram Plot")
plt.show()
```

PIE PLOT

```python
x=[10,20,30,40]
y=["c","c++","Java","Python"]
ex=[0.4,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",shadow=True,startangle=90,counterclock=False)
plt.show()
```

STEM PLOT

```python
x = [1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.stem(x,y,linefmt=":",basefmt="g")

plt.show()
```

BOX PLOT

```python
x = [10,20,30,40,50,60,70,120]

plt.boxplot(x,labels=["python"],showmeans=True,sym="r")
plt.show()
```

STACK PLOT

```python
x = [1,2,3,4,5]
area= [1,2,3,4,5]

plt.stackplot(x,area,colors=["r"])
plt.show()
```

```python
x = [1,2,3,4,5]
area1= [2,3,2,5,4]
area2=[2,3,4,5,6]
area3=[1,3,2,4,2]
l=["area1","area2","area3"]

plt.stackplot(x,area1,area2,area3,labels=l)
plt.legend()
plt.grid()
plt.show()
```
